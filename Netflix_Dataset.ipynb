```python
# Convert duration to minutes
def convert_duration(duration):
    if 'h' in duration:
        parts = duration.replace("h", "").split(" ")
        hours = int(parts[0])
        minutes = int(parts[1]) if len(parts) > 1 else 0
        return hours * 60 + minutes
    elif 'm' in duration:
        # Extract numeric part from duration string
        numeric_duration = ''.join(filter(str.isdigit, duration))
        return int(numeric_duration)
    return np.nan
df['Duration_Minutes'] = df['Duration'].apply(convert_duration)
```

```python
# Convert 'Release_Date' to datetime objects
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df.groupby(df.Release_Date.dt.year)["Show_Id"].count().sort_values(ascending=False)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
import matplotlib.pyplot as plt
# Select and preprocess features
df_sub = df[['Type', 'Rating']].dropna()
# Encode categorical features
for col in ['Type', 'Rating']:
    df_sub[col] = LabelEncoder().fit_transform(df_sub[col].astype(str))
# Scale features
X = df_sub.values
X_scaled = StandardScaler().fit_transform(X)
# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels_kmeans = kmeans.fit_predict(X_scaled)
# Silhouette score
silhouette_kmeans = silhouette_score(X_scaled, labels_kmeans)
print(f"K-Means Silhouette Score (type & rating): {silhouette_kmeans:.3f}")
# Add cluster labels for visualization
df_sub['KMeans_Cluster'] = labels_kmeans
# Visualization
sns.scatterplot(x='Type', y='Rating', hue='KMeans_Cluster', data=df_sub, palette='Set1', s=60)
plt.title('K-Means Clustering on Type and Rating')
plt.xlabel('Type (encoded)')
plt.ylabel('Rating (encoded)')
plt.legend(title='Cluster')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
# Select and preprocess features
df_sub = df[['Type', 'Rating']].dropna()
# Encode categorical features
for col in ['Type', 'Rating']:
    df_sub[col] = LabelEncoder().fit_transform(df_sub[col].astype(str))
# Scale features
X = df_sub.values
X_scaled = StandardScaler().fit_transform(X)
# Agglomerative clustering
agglo = AgglomerativeClustering(n_clusters=3)
labels_agglo = agglo.fit_predict(X_scaled)
# Silhouette score
silhouette_agglo = silhouette_score(X_scaled, labels_agglo)
print(f"Agglomerative Clustering Silhouette Score (type & rating): {silhouette_agglo:.3f}")
# Add cluster labels for visualization
df_sub['Agglo_Cluster'] = labels_agglo
# Visualization
sns.scatterplot(x='Type', y='Rating', hue='Agglo_Cluster', data=df_sub, palette='Set2', s=60)
plt.title('Agglomerative Clustering on Type and Rating')
plt.xlabel('Type (encoded)')
plt.ylabel('Rating (encoded)')
plt.legend(title='Cluster')
plt.show()
```

```python
# === Label Encoding for Categorical Features ===
# Convert 'Type' and 'Rating' columns from categorical to numerical values using LabelEncoder.
# This is required for clustering algorithms which work on numerical data.
from sklearn.preprocessing import LabelEncoder
# Initialize label encoders
type_le = LabelEncoder()
rating_le = LabelEncoder()
# Fit and transform the 'Type' and 'Rating' columns
df['type_encoded'] = type_le.fit_transform(df['Type'].astype(str))
df['rating_encoded'] = rating_le.fit_transform(df['Rating'].astype(str))
# Print the mappings for reference and interpretability
print("Type mapping:", dict(zip(type_le.transform(type_le.classes_), type_le.classes_)))
print("Rating mapping:", dict(zip(rating_le.transform(rating_le.classes_), rating_le.classes_)))
```

```python
# Which is better?
if silhouette_kmeans > silhouette_agglo:
    print("\nK-Means clustering gives better-defined clusters for this dataset based on the silhouette score.")
elif silhouette_kmeans < silhouette_agglo:
    print("\nAgglomerative clustering gives better-defined clusters for this dataset based on the silhouette score.")
else:
    print("\nBoth clustering methods perform equally well based on the silhouette score.")
```
